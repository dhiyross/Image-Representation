# Siapkan library yang akan digunakan

```python
import cv2
import matplotlib.pyplot as plt
```

# Baca representasi citra

```python
# img_bgr = cv2.imread("/content/Bacteria detection with darkfield microscopy_034.png")
img_bgr = cv2.imread("/content/Butterfly Image Classification 75 species_5.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb, vmin=0, vmax=255)
```

# Citra 3 channel

```python
print(img_rgb.shape)
```

```python
img_r = img_rgb[:,:,0]
img_g = img_rgb[:,:,1]
img_b = img_rgb[:,:,2]
```

```python
f, axis = plt.subplots(1,3, figsize=[10, 30])
axis[0].imshow(img_r, cmap="Reds", vmin=0, vmax=255)
axis[1].imshow(img_g, cmap="Greens", vmin=0, vmax=255)
axis[2].imshow(img_b, cmap="Blues", vmin=0, vmax=255)
plt.figure()
plt.imshow(img_rgb)
```

# Konversi color space citra

```python
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
img_lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)
```

```python
plt.figure()
plt.imshow(img_hsv)
plt.figure()
plt.imshow(img_lab)
```

```python
img_hsv[:,:,0].max()
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray", vmin=0, vmax=255)
```

# Histogram

```python
hist = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
hist_r = cv2.calcHist([img_r], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([img_g], [0], None, [256], [0, 256])
hist_b = cv2.calcHist([img_b], [0], None, [256], [0, 256])

# f, axis = plt.subplots(1,3, figsize=[20, 5])
# axis[0].plot(hist_r)
# axis[0].plot(hist_g)
# axis[0].plot(hist_b)

plt.plot(hist_r, color="red")
plt.plot(hist_g, color="green")
plt.plot(hist_b, color="blue")
```
